```python
df.head(1)
```

```python
bumrah_df = df[df.bowler =='JJ Bumrah']
```

```python
bumrah_df.head()
```

```python
# runs scored by this batsman
bdf1 = pd.DataFrame(bumrah_df.groupby('batsman')['batsman_runs'].sum()).reset_index()
```

```python
bdf1.head()
```

```python
bdf2 = pd.DataFrame(bumrah_df.groupby('batsman')['ball'].count()).reset_index()
```

```python
bdf2.head()
```

```python
bdf3 = bdf1.merge(bdf2, on = 'batsman', how = 'left')
```

```python
bdf3.head(1)
```

```python
bdf3['strike_rate'] = 100*bdf3['batsman_runs']/bdf3['ball']
```

```python
bdf3.head(1)
```

```python
# min criteria: 20 balls
```

```python
bdf3 = bdf3[bdf3.ball >= 20]
```

```python
bdf3.head()
```

```python
abd_df = df[df.batsman =='AB de Villiers']
```

```python
# runs scored by this batsman
adf1 = pd.DataFrame(abd_df.groupby('bowler')['batsman_runs'].sum()).reset_index()
```

```python
adf1.head()
```

```python
# balls faced
adf2 = pd.DataFrame(abd_df.groupby('bowler')['ball'].count()).reset_index()
```

```python
adf2.head()
```

```python
adf3 = adf1.merge(adf2, on = 'bowler', how = 'right')
```

```python
adf3.head(1)
```

```python
adf3['strike_rate'] = 100*adf3['batsman_runs']/adf3['ball']
```

```python
adf3.head()
```

```python
# min criteria : 20 balls
```

```python
adf3 = adf3[adf3.ball >= 20]
```

```python
# bdf3, adf3
```

```python
bdf3.head()
```

```python
bdf3.reset_index(inplace = True, drop = True)
adf3.reset_index(inplace = True, drop = True)
```

```python
bdf3.head()
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(16, 8))

# Scatter plot with different colors for 'AB de Villiers'
for i in range(len(bdf3)):
    if bdf3.iloc[i]['batsman'] == 'AB de Villiers':
        plt.scatter(
            bdf3.iloc[i]['strike_rate'],
            bdf3.iloc[i]['batsman_runs'],
            color='blue',  # Set marker color to blue
            edgecolor='red',  # Set edge color to red for highlight
            s=100,  # Marker size
            zorder=5  # Ensure it's on top of other markers
        )
        plt.text(
            bdf3.iloc[i]['strike_rate'],
            bdf3.iloc[i]['batsman_runs'] - 5,  # Adjust position below the point
            bdf3.iloc[i]['batsman'],
            color='red',  # Set the text color to red
            bbox=dict(facecolor='white', edgecolor='red', boxstyle='round,pad=0.5'),  # Add a box around the text
            ha='center'  # Center-align the text horizontally
        )
    else:
        plt.scatter(
            bdf3.iloc[i]['strike_rate'],
            bdf3.iloc[i]['batsman_runs'],
            color='black',  # Set marker color for other batsmen
            edgecolor='none',  # No edge color for other markers
            s=50  # Marker size for other batsmen
        )
        plt.text(
            bdf3.iloc[i]['strike_rate'] + 1,
            bdf3.iloc[i]['batsman_runs'] - 1,
            bdf3.iloc[i]['batsman']
        )

# Add vertical and horizontal lines
plt.axvline(120, ls='--', color='grey')
plt.axhline(60, ls='--', color='grey')

# Set plot title and labels
plt.title('Batsman against Bumrah in IPL (min 20 balls faced)', fontsize=20)
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')

# Show the plot
plt.show()


```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(16, 8))
plt.rcParams['axes.facecolor'] = '#fff7fb'
plt.scatter(adf3['strike_rate'], adf3['batsman_runs'])

# Annotate all bowlers with their names
for i in range(len(adf3)):
    if adf3.iloc[i]['bowler'] == 'JJ Bumrah':
        plt.text(
            adf3.iloc[i]['strike_rate'] + 2,
            adf3.iloc[i]['batsman_runs'] - 1,
            adf3.iloc[i]['bowler'],
            color='maroon'
        )
    elif adf3.iloc[i]['bowler'] in ['RA Jadeja', 'SL Malinga', 'Sandeep Sharma', 'Harbhajan Singh']:
        plt.text(
            adf3.iloc[i]['strike_rate'] + 2,
            adf3.iloc[i]['batsman_runs'] - 1,
            adf3.iloc[i]['bowler']
        )
    elif adf3.iloc[i]['bowler'] == 'HH Pandya':
        plt.text(
            adf3.iloc[i]['strike_rate'],
            adf3.iloc[i]['batsman_runs'] + 2,
            adf3.iloc[i]['bowler']
        )
    elif adf3.iloc[i]['bowler'] == 'AD Russell':
        plt.text(
            adf3.iloc[i]['strike_rate'] - 3,
            adf3.iloc[i]['batsman_runs'] - 3,
            adf3.iloc[i]['bowler']
        )

# Add vertical and horizontal lines
plt.axvline(110, ls='--', color='grey')
plt.axhline(80, ls='--', color='grey')

# Set plot title and labels
plt.title("ABD's Favorite Bowlers", fontsize=20)
plt.xlabel('Strike Rate')
plt.ylabel('Runs Scored')

# Save and show the plot
plt.savefig('Chart_ABD_vs_Bumrah_part2.jpg')
plt.show()

```
